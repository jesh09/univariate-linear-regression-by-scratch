```python
x = data['x']
y = data['y']
print(x.shape,y.shape)
```

```python
def line(m,x):
    return np.dot(m,x)
```

```python
def error(m,x,y):
    return np.mean(((np.dot(m,x))-y)**2)
```

```python
def slope(m,x,y):
    return np.mean(2*((np.dot(m,x))-y)*x)

```

```python
def train(lr,m,x,y,iters):
    for i in range(iters):
        
        m = m - lr*slope(m,x,y)
        
        #c = c - lr*diffIntercept(m,c,x,y)
        return [m]
```

```python
params = train(0.000000001,0.5,x,y,100)
```
